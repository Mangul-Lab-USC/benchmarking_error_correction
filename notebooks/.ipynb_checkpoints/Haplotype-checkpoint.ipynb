```python
from __future__ import division
import pylab as pl
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import math
import seaborn as sns
import os
```

# Mix datasets

```python
data = pd.read_csv('../summary_data/haplotype_summary.csv')
print(data.columns)
data["Trim Effeciency"] = data["Base - TP TRIM"]/(data["Base - TP TRIM"] + data["Base - FP TRIM"])
data['total_corrections']=data['Base - TP']+ data['Base - FP']
data = data[(data['Kmer Size'] >19)]
data = data[(data['Kmer Size'] <= 30)]
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
g=sns.set_style("white")
g=sns.set_context("talk")
hap=1
g = sns.catplot(x="Kmer Size", y="Base Gain", kind="point",data = data[(data['hap'] ==hap)]
, hue='Tool')
g.set(xlabel='Kmer size', ylabel='Gain', title="Hap "+str(hap))
g=sns.despine()
plt.savefig("../figures/haplotype_gain_hap"+str(hap)+".png", bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
g=sns.set_style("white")
g=sns.set_context("talk")

g=sns.catplot(x="total_corrections", y="Tool" , col = "hap", col_wrap = 3,data = data[(data['Kmer Size'] >= 30)], kind = "bar")
g.set(xlabel='Total number of corrections', ylabel='Methods', xscale='log')
g.fig.tight_layout()
g=sns.despine()
plt.show(g)


plt.savefig("../figures/haplotype_total_corrections.pdf", bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
g=sns.set_style("white")
g=sns.set_context("talk")
g = sns.catplot(x="Error", y="Base - FP", kind="point",data=data,hue='Tool')
, hue='Tool')
g.set(xlabel='Error', ylabel='FP')
g.set_xticklabels(rotation=36)
g=sns.despine()
#plt.show(g)
plt.savefig("../figures/mix_FP_vs_error.png", bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
g=sns.set_style("white")
g=sns.set_context("talk")

result = data.pivot(index='Tool', columns='hap', values='Base Gain')

g=sns.heatmap(result, fmt="g", cmap='gist_ncar',center=0,linewidths=.5)
plt.savefig("../figures/haplotype_heatmap_gain_vs_hap.png", bbox_inches='tight')
```

```python
result_PPV = data.pivot(index='Tool', columns='Error', values='Base Precision')

import matplotlib.pyplot as plt
import seaborn as sns
g=sns.set_style("white")
g=sns.set_context("talk")
g=sns.heatmap(result_PPV, fmt="g", cmap='gist_ncar',center=0,linewidths=.5)
plt.savefig("../figures/mix_heatmap_ppv_vs_error.png", bbox_inches='tight')
```

```python
result_Sensitivity = data.pivot(index='Tool', columns='Error', values='Base Sensitivity')

import matplotlib.pyplot as plt
import seaborn as sns
g=sns.set_style("white")
g=sns.set_context("talk")
g=sns.heatmap(result_Sensitivity, fmt="g", cmap='gist_ncar',center=0,linewidths=.5)
plt.savefig("../figures/mix_heatmap_sensitivity_vs_error.png", bbox_inches='tight')
```
